```python
#V01 12:31
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#打开文件
df1 = pd.read_excel(r'D:\#条形码系统\盘点\库存对照\20200113\《资材》材料在库数据_202001130800.xls',usecols=[1,4,5,6,31])
df2 = pd.read_excel(r'D:\#条形码系统\盘点\库存对照\20200113\库存汇总表202001130751.xlsx',usecols=[0,3,6])
df2 = pd.pivot_table(df2,values=['期末数量'],index='资材编号',aggfunc='sum').reset_index()
df3 = pd.read_excel(r'D:\#条形码系统\盘点\库存对照\20200113\数据清洗20170110.xlsx',sheet_name='在庫',usecols=[1,3])
df4 = pd.read_excel(r'D:\#条形码系统\盘点\库存对照\20200113\物料信息列表.xlsx',usecols=[0,3])
#连接表，比较各字段内容是否相同
df = pd.merge(df1,df4,on='资材编号',how='left')
df = pd.merge(df,df2,on='资材编号',how='left')
df = pd.merge(df,df3,on='资材编号',how='left')
df = df.fillna(0)

df['库存差'] = df.apply(lambda x:x.现库存数量-x.期末数量,axis=1)
df[(df['库存差']!=0) & (df['材料担当']=='井海峰')]
```

```python
df.to_excel(
    excel_writer=r'D:\#条形码系统\盘点\库存对照\20200113\库存对照表202001141554.xlsx',
    index=False
)
```
